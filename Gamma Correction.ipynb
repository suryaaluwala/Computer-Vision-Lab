```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a sample image
img = cv2.imread('butterfly.jpg')

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Normalize pixel values (0-1)
normalized = img / 255.0


# Gamma value
gamma = 0.5     # less than 1 -> brighter


# Apply gamma correction
gamma_corrected = np.power(normalized, gamma)


# Convert back to 0-255 range
gamma_corrected = np.uint8(gamma_corrected * 255)


# Display original and corrected images
result = np.hstack((img, gamma_corrected))

plt.figure(figsize=(12,5))
plt.imshow(result)
plt.title("Original | Gamma Corrected")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Gamma values
gammas = [0.5, 1.0, 2.0]

images = []

for gamma in gammas:

    # Normalize
    normalized = img / 255.0

    # Gamma correction
    corrected = np.power(normalized, gamma)

    # Convert back
    corrected = np.uint8(corrected * 255)

    images.append(corrected)


# Stack results
output = np.hstack((img, images[0], images[1], images[2]))


plt.figure(figsize=(16,5))
plt.imshow(output)

plt.title(
    "Original | Gamma 0.5 (Bright) | Gamma 1 | Gamma 2 (Dark)"
)

plt.axis('off')
plt.show()
```
