```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read images
img1 = cv2.imread('butterfly.jpg')
img2 = cv2.imread('tree.jpg')

# Resize second image to match first
img2 = cv2.resize(
    img2,
    (img1.shape[1], img1.shape[0])
)

# Blend images
blended = cv2.addWeighted(
    img1,
    0.5,   # weight of first image
    img2,
    0.5,   # weight of second image
    0
)

# Display
result = np.hstack((img1, img2, blended))

plt.figure(figsize=(15,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Image1 | Image2 | Blended")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('butterfly.jpg')

# Create second image
img2 = cv2.flip(img1, 1)

# Blend
blended = cv2.addWeighted(
    img1,
    0.5,
    img2,
    0.5,
    0
)

result = np.hstack((img1, img2, blended))

plt.figure(figsize=(15,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Flipped | Blended")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('butterfly.jpg')
img2 = cv2.flip(img1, 1)

blend1 = cv2.addWeighted(img1, 0.8, img2, 0.2, 0)
blend2 = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
blend3 = cv2.addWeighted(img1, 0.2, img2, 0.8, 0)

result = np.hstack((
    blend1,
    blend2,
    blend3
))

plt.figure(figsize=(15,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("0.8:0.2 | 0.5:0.5 | 0.2:0.8")
plt.axis('off')
plt.show()
```
