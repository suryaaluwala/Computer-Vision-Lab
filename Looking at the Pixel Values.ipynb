```python
import cv2
import pandas as pd

# Read image
img = cv2.imread('butterfly.jpg')

# Check image loaded
if img is None:
    print("Image not found")
else:
    

    # Convert image pixels into rows
    pixels = img.reshape(-1, 3)

    # Create DataFrame with BGR values
    df = pd.DataFrame(
        pixels,
        columns=['Blue', 'Green', 'Red']
    )

    # Save to Excel
    df.to_excel(
        'pixel_values.xlsx',
        index=False
    )

    print("Pixel values saved to Excel")
```
