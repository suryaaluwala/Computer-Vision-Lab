```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Convert BGR to RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Color space conversions
hsv  = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
xyz  = cv2.cvtColor(rgb, cv2.COLOR_RGB2XYZ)
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
yuv  = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)
lab  = cv2.cvtColor(rgb, cv2.COLOR_RGB2LAB)

# Convert gray back to RGB for stacking
gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

# Stack images
result = np.hstack((rgb, hsv, xyz, gray_rgb, yuv, lab))

# Display
plt.figure(figsize=(20,5))
plt.imshow(result)
plt.title("RGB | HSV | XYZ | GRAY | YUV | LAB")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Convert BGR to RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# RGB -> HSV -> RGB
hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
rgb_hsv = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

# RGB -> XYZ -> RGB
xyz = cv2.cvtColor(rgb, cv2.COLOR_RGB2XYZ)
rgb_xyz = cv2.cvtColor(xyz, cv2.COLOR_XYZ2RGB)

# RGB -> LAB -> RGB
lab = cv2.cvtColor(rgb, cv2.COLOR_RGB2LAB)
rgb_lab = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)

# RGB -> YUV -> RGB
yuv = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)
rgb_yuv = cv2.cvtColor(yuv, cv2.COLOR_YUV2RGB)

# RGB -> Gray -> RGB
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
rgb_gray = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

# Stack horizontally
result = np.hstack((
    rgb,
    rgb_hsv,
    rgb_xyz,
    rgb_lab,
    rgb_yuv,
    rgb_gray
))

# Display
plt.figure(figsize=(24,6))
plt.imshow(result)
plt.title("Original RGB | HSV→RGB | XYZ→RGB | LAB→RGB | YUV→RGB | GRAY→RGB")
plt.axis('off')
plt.show()
```
