```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Increase brightness
bright = np.clip(img + 50, 0, 255)

# Display
result = np.hstack((img, bright))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Brightness Increased")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

# Increase contrast
contrast = np.clip(1.8 * img, 0, 255).astype(np.uint8)

result = np.hstack((img, contrast))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Contrast Increased")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

alpha = 1.5   # Contrast
beta = 40     # Brightness

adjusted = np.clip(alpha * img + beta, 0, 255).astype(np.uint8)

result = np.hstack((img, adjusted))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Adjusted")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

bright = np.clip(img + 50, 0, 255).astype(np.uint8)

dark = np.clip(img - 50, 0, 255).astype(np.uint8)

contrast = np.clip(1.8 * img, 0, 255).astype(np.uint8)

adjusted = np.clip(1.5 * img + 40, 0, 255).astype(np.uint8)

result = np.hstack((
    img,
    bright,
    dark,
    contrast,
    adjusted
))

plt.figure(figsize=(20,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title(
    "Original | Bright | Dark | Contrast | Adjusted"
)
plt.axis('off')
plt.show()
```
