```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('surya.jpg')
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis('off')
plt.show() 
print(img.shape)
# Saving of image 
cv2.imwrite("surya.jpg",img)
print("Task is Done !")


# Printing of Pixel Values
#print(img)



# Saving to CSV
  # converting img pixels into rows
pixels = img.reshape(-1,3)
#Creating of dataframe with BGR values
df = pd.DataFrame(
    pixels,
    columns = ['B','G','R']
)
#now saving to Excel
df.to_csv(
    'pixel_values.csv',
    index=False
)
print("The Excel Task is Done !")



#img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
xyz = cv2.cvtColor(rgb, cv2.COLOR_RGB2XYZ)
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
yuv = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)
lab = cv2.cvtColor(rgb, cv2.COLOR_RGB2LAB)

# converting gray to rgb (bcz of it would be in 2D ) to print it along with above serial output.
gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

# printing of all diff formats in same line ouput
res = np.hstack((rgb,hsv,xyz,gray_rgb,yuv,lab))

plt.figure(figsize=(20,5))
plt.imshow(res)
plt.axis('off')
plt.show()
```
