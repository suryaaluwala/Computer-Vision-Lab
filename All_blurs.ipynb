```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("saul.jpg")

# Resize so stacking works properly
img = cv2.resize(img, (300, 300))

# Different blurs
gaussian = cv2.GaussianBlur(img, (15, 15), 0)
median = cv2.medianBlur(img, 5)
bilateral = cv2.bilateralFilter(img, 9, 75, 75)
box = cv2.blur(img, (15, 15))

# Stack images horizontally
stack = np.hstack((img, gaussian, median, bilateral, box))

# Convert BGR → RGB for display
stack_rgb = cv2.cvtColor(stack, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15, 5))
plt.imshow(stack_rgb)
plt.axis("off")
plt.title("Original | Gaussian | Median | Bilateral | Box Blur")
plt.show()
```
