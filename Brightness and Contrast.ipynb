```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Increase brightness
bright = cv2.convertScaleAbs(
    img,
    alpha=1,
    beta=50
)

# Display
result = np.hstack((img, bright))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Brightness Increased")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Increase contrast
contrast = cv2.convertScaleAbs(
    img,
    alpha=2,
    beta=0
)

# Display
result = np.hstack((img, contrast))

plt.figure(figsize=(12,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Original | Contrast Increased")
plt.axis('off')
plt.show()
```
