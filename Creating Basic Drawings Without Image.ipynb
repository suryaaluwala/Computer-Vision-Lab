```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black blank image
img = np.zeros((500, 500, 3), dtype=np.uint8)

# Draw a dot
# center=(x,y), radius, color(B,G,R), thickness=-1 for filled
cv2.circle(img, (250,250), 10, (0,0,255),-1)

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Blank canvas
img = np.zeros((500,500,3), dtype=np.uint8)

# Points
points = [(100,100), (200,200), (300,300), (400,400)]

# Draw dots
for p in points:
    cv2.circle(img, p, 15, (0,255,0), -1)

# Display
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create blank black image
img = np.zeros((500, 700, 3), dtype=np.uint8)

# 1. Draw a Line
cv2.line(
    img,
    (50,50),        # starting point
    (300,50),       # ending point
    (0,255,0),      # green color (BGR)
    5               # thickness
)

# 2. Draw a Rectangle
cv2.rectangle(
    img,
    (50,100),       # top-left
    (250,250),      # bottom-right
    (255,0,0),      # blue
    3
)

# 3. Draw Filled Rectangle
cv2.rectangle(
    img,
    (300,100),
    (500,250),
    (0,0,255),
    -1              # filled
)

# 4. Draw Circle
cv2.circle(
    img,
    (150,350),      # center
    70,             # radius
    (0,255,255),
    4
)

# 5. Draw Dot
cv2.circle(
    img,
    (400,350),
    10,
    (255,255,255),
    -1
)

# 6. Add Text
cv2.putText(
    img,
    "OpenCV",
    (450,80),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (255,255,255),
    2
)

# 7. Draw Polygon
points = np.array([
    [500,300],
    [600,300],
    [650,450],
    [550,450]
])

cv2.polylines(
    img,
    [points],
    True,
    (0,255,0),
    4
)

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.figure(figsize=(10,7))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((400,400,3), dtype=np.uint8)

# Draw ellipse
cv2.ellipse(
    img,
    (200,200),
    (120,60),
    0,
    0,
    360,
    (255,0,255),
    4
)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
