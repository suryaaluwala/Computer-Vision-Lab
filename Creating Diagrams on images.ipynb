```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Draw a line
# cv2.line(image, start_point, end_point, color, thickness)
cv2.line(img, (50,50), (300,200), (0,255,0), 5)

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

# Points of polygon
points = np.array([
    [100,100],
    [300,50],
    [400,200],
    [200,300]
])

# Draw polygon
cv2.polylines(
    img,
    [points],
    True,
    (0,255,0),
    4
)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

line = img.copy()
rect = img.copy()
circle = img.copy()
text = img.copy()

cv2.line(line,(50,50),(300,200),(0,255,0),5)

cv2.rectangle(rect,(100,100),(400,300),(255,0,0),3)

cv2.circle(circle,(250,200),80,(0,255,255),4)

cv2.putText(text,"Butterfly",
            (50,50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,(255,255,255),2)

# Stack
result = np.hstack((line,rect,circle,text))

plt.figure(figsize=(20,5))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('butterfly.jpg')

# Draw a dot
# (image, center, radius, color, thickness)
cv2.circle(img, (200,150), 5, (0,0,255), -1)

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('butterfly.jpg')

# Draw multiple dots
points = [(100,100), (200,150), (300,200), (400,250)]

for point in points:
    cv2.circle(img, point, 8, (0,255,0), -1)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
